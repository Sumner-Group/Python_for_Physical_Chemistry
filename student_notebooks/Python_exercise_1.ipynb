### Naming multiple values

Sometimes, you may want to assign several values to a variable. For example, you may want to save a set of energy eigenvalues for the particle-on-a-line problem. Perhaps the values of the eigenvalues are (in atomic units) 1.0, 4.0, 9.0, 16.0, 25.0, 36.0. You could refer to these eigenvalues one at a time. Alternatively, you could assign these values to a single name and simplify your typing. In doing this, you save yourself some effort and reduce the opportunities for errors.

So, how to we assign multiple values to a name? First, you select your name (ex. `En`), draw your `=`, then type `[]`. The `[]` takes multiple inputs and combines them into a single unit which we name. The values are entered between the `[]` separated by commas. If we take the energy values from the above section and put them into `[]`, our result is:

`En = [1.0, 4.0, 9.0, 16.0, 25.0, 36.0]`

The `En` object can now be used for calculation without having to type and re-type this string every time. The formal name of the `En` object is an **array**. An array requires that all the data be the same type, meaning all numbers or all text and you *cannot* mix data types. (An array in Python is similar to a vector in linear algebra.)

In the previous example above, each value of energy was calculated beforehand and then added into the array. It might be nice instead to have the computer to all of the work. In order to do this, we need to use a structure called a **loop**. In a loop, a calculation is repeated over and over until a condition is met. Let's run the following example:

```python
# create the loop parameters
for n in range(5):
  # do the math
  square = n**2
  # report out the results
  print("n = ",n,"; n**2 = ",square)
```

The above is called a *loop*. These types of loops are used frequently and is the underlying code for when you drag a formula through a column in Excel or Sheets to create a new column of numbers.

Examine the code and the output carefully. The variable `n` loops from `n = 0` to `n = 4`. The code says `for n in range (5):`. So, when we enter the loop, `n` is automatically set to `0` and it carries out the calculation inside the loop 5 times, i.e. `range(5)` .Notice also that the line with the `for` statement ends in a colon, `:`. This is a **requirement**. WIthout the colon, the code will not work.

It is also important to notice the tabs/alignment. **Everything** that happens inside loop is tabbed. If we want to print something outside the loop, it needs to be outside the tab. Run the following example code:


```python
# create the loop parameters
for n in range(5):
  # do the math
  square = n**2
  # report out the results, kg is the number and we follow that with the units
  print("n**2 inside the loop ",square)
print("n**2 outside the loop", square)
```

Notice how `n**2 inside the loop` is printed 5 times but `n**2 outside the loop` is printed once. Also notice the the value of `n**2 outside the loop` is the same as the last value of `n**2 inside the loop`. This is because once `n` gets to its fifth value (remember the first value is `0` and the last value is `4`), the code exits the loop and `n` no longer increases.

Now let's consider how we could create the array `En = [1.0, 4.0, 9.0, 16.0, 25.0 36.0]` using a loop.
1. The first value needs to be `1` and **not** `0`.
2. We want to calculate six values, so we want `range(6)` **not** `range(5)`
3. We want to store all the values in an array. not as individual numbers
---
Let's see how this could work. Read and then run the following example.

```python
#Create an empty array called En. We will fill this later. Basically, this tells
#Python that we plan to create an array and it should save some computer memory for it.
En=[]
# create the loop parameters. Use range(6) since we want 6 values not five
for n in range(6):
  # do the math. We use (n+1) since we want our first value to be 1 and not 0
  square = (n+1)**2
  # Add the value square to the array En
  En.append(square)
# report out the results
print("The energy values are ",En)
```

## NumPy and SciPy

Like we download software or apps onto devices to do new things, we can customize Python with apps, called packages or libraries. These libraries contain pre-packaged functions that can do math or handle data in ways that make it easier for us. The code below tells Python to use this package and that we will designate any commands that use NumPy, with an np. An example of calculating an average is shown.

```python
import numpy as np

# calculate the average of the first six energy eigenvalues
np.mean(En)
```

SciPy is another popular package that we will use. Among its many uses, it contains a list of many important physical constants and conversion factors. This means that if you import SciPy and load the scipy.constants function, you will not have to pre-program values for Planck's constant, mass of the electron, speed of light, etc. An example is shown below. For a list of available SciPy constants, see: https://docs.scipy.org/doc/scipy/reference/constants.html

```python
#import the physical constants package from scipy
from scipy import constants
#print out some values
print("Planck constant = ", constants.Planck)
print("Reduced Planck constant = ", constants.hbar)
print("Electron mass = ",constants.m_e)
print("Avogadro constant = ", constants.N_A)
```
