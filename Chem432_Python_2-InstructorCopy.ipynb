## Making an energy calculator

Let's create a new code. This code will use the particle-on-a-line model to calculate the difference in energy between two adjacent energy levels (ie, the HOMO-LUMO gap).

We should be able to use markdown to input the value of the HOMO and the length of the molecule. The output will be the energy of the HOMO-LUMO gap in kJ/mol and the wavelength of light absorbed.

We will use all of the skills we have thus far built up. Remember, *if there is a blank space after a comment, you need to write some code.*

```python
# Load up NumPy
import numpy as np
# Load up SciPy constants
from scipy import constants
# Add in a markdown title that automatically reruns the code when new parameters are entered
# @title #HOMO-LUMO Gap Calculator { run: "auto" }

# Add in a markdown parameter, L, for length in angstroms
#@markdown Bond Length in Angstroms
L=5.74#@param{type:"number"}

# Add in a markdown parameter, n_homo, for quantum number associated with the HOMO
#@markdown Quantum Number of HOMO
n_homo=2#@param{type:"number"}

# Set the value of Planck's constant using scipy constants. The code below shows how
#to assign the scipy constants to a variable so that you don't have to always type
# constants.something
h = constants.Planck
# Set the value of electron mass in kg
m_e = constants.electron_mass

#set the value of the speed of light
c = constants.speed_of_light

# Calculate the value of the LUMO, n_lumo
n_lumo = n_homo+1

# Convert length into meters
L = L*1E-10

# Calculate the energy of the HOMO, E_homo
E_homo = n_homo**2*h**2/(8*m_e*L**2)

# Calculate the energy of the LUMO, E_lumo
E_lumo = n_lumo**2*h**2/(8*m_e*L**2)

# Calculate the change in energy, DeltaE
DeltaE = E_lumo - E_homo

# Calculate the energy in kJ/mol
DeltaE_kJ = DeltaE/1000*constants.N_A

# print the energy in kJ/mol
print("The energy of the HOMO-LUMO gap is ",DeltaE_kJ,"kJ/mol")

# Calculate the wavelength of light, Lambda
Lambda = h*c/DeltaE

# print the wavelength in nm
print("The wavelength of light absorbed is ",Lambda*1E9,"nm")
```

## Plotting
Next, we will learn a bit about plotting with python. We will plot the 1D particle-on-a-line wavefunctions

We will need to use two libraries: MatPlotLib and NumPy. In the **code chunk** below, if there is a blank line(s) after a comment, you need to fill it in. If there is already a line of code after a comment, that is one that I wrote for you and you don't have to write anything.

```python
# Add in a markdown title that automatically reruns the code when new parameters are entered
# @title #1D Particle-on-a-line Wavefunctions { run: "auto" }
# Load in NumPy
import numpy as np
# Load in MatPlotLib
import matplotlib.pyplot as plt
# Add in a markdown paramter, L, for length in angstroms
#@markdown Bond Length in Angstroms
L=5.74#@param{type:"number"}

# Add in a markdown parameter, n, for quantum number associated with the wavefunction you'd like to print
#@markdown Quantum Number
n=2#@param{type:"number"}

# Create an NumPy array called x that contains 100 evenly spaced values from 0 to L.
# These are your x-values, or x-axis.
x=np.linspace(0,L,100)
# Create your wavefunction, psi. Be sure to make it normalized.
# Also, use np.sqrt() for square root, np.sin() for sine, and np.pi for pi.
psi=np.sqrt(2/L)*np.sin(n*np.pi*x/L)

# Plot x on the x-axis and psi on the y-axis.
plt.plot(x,psi)
# Set the x-limits of the plot. Plot from 0 to L.
plt.xlim(0,L)
# Give your plot a title. Make the font 16pt.
plt.title("Particle in Box",fontsize=16)
# Label the y-axis. Call it psi_n(x), where n changes depending on what value you want to plot
plt.ylabel("$\\psi_%i$(x)" %n,fontsize=16)
# Label the x-axis with the Angstrom symbol.
plt.xlabel("($\\AA$)",fontsize=16)
# Show the plot
plt.show()
```
